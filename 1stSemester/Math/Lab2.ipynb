```python
# Program 1: Program to plot curve connecting points using plot and scatter modules

from pylab import *
x = [2, 3, 4, 5, 7]
y = [4, 2, 3, 7, 9]
if len(x) != len(y):
    print("Number of values of x and y does not match!")
else:
    scatter(x, y) # Marks only the points
    plot(x, y) # Connects the points and plots the graph
    grid() # Displays a grid for visual assistance
    xlabel("x-axis")
    ylabel("y-axis")
    title("Curve connecting the given points")
    axvline(x = 0, color = "red") # Constructs a vertical line for y = 0 (x axis)
    axhline(y = 0, color = "red") # Constructs a horizontal line for x = 0 (y axis)
    show () # idk why this is required cuz the output is shown even without this line.. but whatever...
    
```

```python
# Program 2: Program to plot exponential function y = e^x
from pylab import *
x = arange(-5, 5, 0.001) # used to generate an array of values for the variable x over the specified range, from -5 to 5, with a step of 0.001
y = exp(x)
plot(x, y) # Plots the graph
title("$y = e^x$")
grid()
xlabel("x-axis")
ylabel("y-axis")
xlim(-5, 5)
ylim(-1, 20)
axvline(x = 0, color = "red")
axhline(y = 0, color = "red")
show()
```

```python
# Program 3: Program to plot sin(x) and cos(x) together
from pylab import *
x = linspace(-2 * pi, 2 * pi, 100) # Generates an array of evenly spaced values within a specified interval

plot(x, sin(x), label = "$y = sin(x)$")
plot(x, cos(x), label = "$y = cos(x)$")

title("Exponential Curve")
grid()
xlabel("x-axis")
ylabel("y-axis")
title("sin(x) and cos(x)")
legend() # Displays the Legend (Identification of lines)

xlim(-2 * pi, 2 * pi)
axvline(color = "red")
axhline(color = "red")
show()
```

```python
# Program 4: Program to plot four leaved rose r = 2 |cos(2 * theta)|

from pylab import * 
theta = linspace(0, 2 * pi, 1000)
r = 2 * abs(cos(2 * theta))
polar(theta, r) # Used to create a polar plot, which represents data in terms of angle (theta) and radius (r) instead of the typical Cartesian x-y coordinates.
title("$r = 2 |cos(2 * \\theta)|$")
axvline(color = "red")
show()
```

```python
# Exercise 2: Write python program to plot linear curve y = x, quadratic curve y = x^2 and cubic curve y = x^3

from pylab import *
n = float(input("Enter the value of n: "))

x = arange(-n, n, 0.1)
y1 = x
y2 = x ** 2
y3 = x ** 3

plot(x, y1, label = "x")
plot(x, y2, label = "$x^2$")
plot(x, y3, label = "$x^3$")

xlabel("X-axis")
ylabel("Y-axis")
title("$y = x, y = x^2, y = x^3$")
grid()
legend(loc = "upper right")
show()
```

```python
# Exercise 3: Write python program to plot cardioids r = a(1 + cos(theta)) and r = a(1 - cos(theta)) taking a = 2

from pylab import *

theta = linspace(0, 2 * pi, 1000)
r1 = 2 * (1 + cos(theta))
r2 = 2 * (1 - cos(theta))

polar(theta, r1, theta, r2)

title("Cardiods: $r = a(1 + cos(\\theta))$ and $r = a(1 - cos(\\theta))$")
axvline(color = "red")
show()
```
